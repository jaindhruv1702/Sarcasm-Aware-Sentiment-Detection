```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.util import ngrams
import string
import re

# Input text
text = """
Wow!!! NLP is amazing 😊😊
Visit https://example.com for more info.
Email us at support@abc.com
Price is 500₹ only!!!
#AI #NLP
"""

print("Original Text:")
print(text)

text = text.lower()
text = re.sub(r'https\S+|www\S+', '', text)
text = re.sub(r'\S+@\S+', '',text)
text = re.sub(r'[^\x00-\x7F]+', '', text)
text = text.replace('#', '')

print("After Filterartion Text:")
print(text)



# 1. Tokenization

tokens = word_tokenize(text)

print("\nTokens:")
print(tokens)

# 1. Filtration

stop_words = set(stopwords.words('english'))

filtered_words = []

for word in tokens:
  if word not in stop_words and word not in string.punctuation:
    filtered_words.append(word)
print("\nFiltered Words:")
print(filtered_words)
```

```python
#Tokenization

tokens = word_tokenize(text)
print('\n After Tokenization: ')
print(tokens)

#unigram
unigrams = list(ngrams(tokens,1))
uni_freq = Counter(unigrams)

print('Unigrams: ')
print(unigrams)

total_words = len(tokens)
#probabilities

print("\nUnigram Probabilities:\n")
for word, count in uni_freq.items():
  probability = count/total_words
  print(f"P{word} = {count}/{total_words} = {probability:.3f}")

# print(total_words)
# print(uni_freq)

#bigram
bigrams = list(ngrams(tokens, 2))

bigram_freq = Counter(bigrams)

print("\nBigrams:")
print(bigram_freq)

# Unigram counts for denominator
word_counts = Counter(tokens)

print("\nBigram Probabilities:\n")

for bigram, count in bigram_freq.items():

    first_word = bigram[0]

    probability = count / word_counts[first_word]

    print(f"P({bigram[1]} | {bigram[0]}) = "
          f"{count}/{word_counts[first_word]} = "
          f"{probability:.3f}")

# print(bigrams)
# print(bi_freq)

#Trigram

trigram = list(ngrams(tokens,3))
print("Trigrams:")
print(trigram)

tri_freq = Counter(trigram)

bigram_counts = Counter(bigrams)

print('\nTrigram Probabilities: ')

for trigram, count in tri_freq.items():
  first_bigram = (trigram[0], trigram[1])

  probability = count / bigram_counts[first_bigram]

  print(f"P({trigram[2]} | {trigram[0]}, {trigram[1]}) = "
          f"{count}/{bigram_counts[first_bigram]} = "
          f"{probability:.3f}")
```

# **Prac 5**

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.metrics.distance import edit_distance

# Input text
text = "I am lernning naturall language processsing"

print("Original Text:")
print(text)

#Tokenization
tokens = word_tokenize(text)
print("\n After Tokenization")
print(tokens)

#Stop word removal
stop_words = set(stopwords.words('english'))

filtered_words = []

for word in tokens:
  if word.lower() not in stop_words:
    filtered_words.append(word)

print("\nAfter Stop Word Removal:")
print(filtered_words)

# Stemming
ps = PorterStemmer()

stemmed_words = []

for word in filtered_words:
  stemmed_words.append(ps.stem(word))

print("\nStemmed Words:")
print(stemmed_words)

# Spell correction

correct_words = ["learning",
    "natural",
    "language",
    "processing"]


for word in filtered_words:

    min_distance = float('inf')
    correction = ''

    for correct_word in correct_words:

        distance = edit_distance(word, correct_word)

        if distance < min_distance:
            min_distance = distance
            correction = correct_word

    print(f"\nMisspelled Word: {word}")
    print(f"Suggested Correction: {correction}")
    print(f"Edit Distance: {min_distance}")
    print("-----------------------")
```

```python
import spacy
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords

# Download NLTK resources
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('stopwords')

# Load spaCy model
nlp = spacy.load("en_core_web_sm")

# Resume text
text = """
Dhruv Jain is a Computer Science student from Mumbai.
He completed an internship at TCS.
He knows Python, Machine Learning, NLP, and Java.
"""

print("Original Resume Text:\n")
print(text)

# --------------------------------
# 1. Tokenization
# --------------------------------

tokens = word_tokenize(text)

print("\nTokens:\n")
print(tokens)

# --------------------------------
# 2. Stop-word Removal
# --------------------------------

stop_words = set(stopwords.words('english'))

filtered_words = []

for word in tokens:

    if word.lower() not in stop_words:
        filtered_words.append(word)

print("\nAfter Stop-word Removal:\n")
print(filtered_words)

# --------------------------------
# 3. Lemmatization
# --------------------------------

doc = nlp(text)

lemmatized_words = []

for token in doc:
    lemmatized_words.append(token.lemma_)

print("\nLemmatized Words:\n")
print(lemmatized_words)

# --------------------------------
# 4. Named Entity Recognition
# --------------------------------

print("\nNamed Entities:\n")

for ent in doc.ents:

    print(f"Entity: {ent.text}")
    print(f"Label : {ent.label_}")
    print("----------------------")
```

# **Prac 7**

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.util import ngrams
from collections import Counter

# Download resources
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('stopwords')

# Training text
text = """
I love NLP
I love machine learning
I love Python
NLP is fun
Machine learning is powerful
"""

print("Original Text:\n")
print(text)

# --------------------------------
# 1. Tokenization
# --------------------------------

tokens = word_tokenize(text.lower())

print("\nTokens:\n")
print(tokens)

# --------------------------------
# 2. Stop-word Removal
# --------------------------------

stop_words = set(stopwords.words('english'))

filtered_words = []

for word in tokens:

    if word not in stop_words:
        filtered_words.append(word)

print("\nFiltered Words:\n")
print(filtered_words)

# --------------------------------
# 3. Stemming
# --------------------------------

ps = PorterStemmer()

stemmed_words = []

for word in filtered_words:
    stemmed_words.append(ps.stem(word))

print("\nStemmed Words:\n")
print(stemmed_words)

# --------------------------------
# 4. Bigram Model
# --------------------------------

bigrams = list(ngrams(stemmed_words, 2))

print("\nBigrams:\n")
print(bigrams)

# Frequency count
bigram_freq = Counter(bigrams)

# --------------------------------
# 5. Next Word Prediction
# --------------------------------

input_word = "love"

print(f"\nNext word predictions for '{input_word}':\n")

predictions = []

for bigram in bigram_freq:

    if bigram[0] == input_word:

        predictions.append((bigram[1], bigram_freq[bigram]))

# Sort by frequency
predictions.sort(key=lambda x: x[1], reverse=True)

for word, freq in predictions:

    print(f"Next Word: {word} | Frequency: {freq}")

# --------------------------------
# 6. Sentence Probability
# --------------------------------

print("\nSentence Probability:\n")

word_counts = Counter(stemmed_words)

for bigram, count in bigram_freq.items():

    probability = count / word_counts[bigram[0]]

    print(f"P({bigram[1]} | {bigram[0]}) = "
          f"{count}/{word_counts[bigram[0]]} = "
          f"{probability:.3f}")
```

# **Prac 8**

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.classify import NaiveBayesClassifier

# Download resources
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('stopwords')

# --------------------------------
# Training Data
# --------------------------------

training_data = [

    ("I deposited money in the bank", "finance"),

    ("She withdrew cash from the bank", "finance"),

    ("The river bank is beautiful", "river"),

    ("Children are playing near the bank", "river")
]

# --------------------------------
# Stopwords
# --------------------------------

stop_words = set(stopwords.words('english'))

# --------------------------------
# Feature Extraction Function
# --------------------------------

def extract_features(sentence):

    words = word_tokenize(sentence.lower())

    filtered_words = []

    for word in words:

        if word not in stop_words:

            filtered_words.append(word)

    features = {}

    for word in filtered_words:

        features[word] = True

    return features

# --------------------------------
# Prepare Training Set
# --------------------------------

training_set = []

for sentence, label in training_data:

    features = extract_features(sentence)

    training_set.append((features, label))

# --------------------------------
# Train Naïve Bayes Classifier
# --------------------------------

classifier = NaiveBayesClassifier.train(training_set)

# --------------------------------
# Test Sentence
# --------------------------------

test_sentence = "He went to the bank to deposit money"

test_features = extract_features(test_sentence)

# --------------------------------
# Prediction
# --------------------------------

prediction = classifier.classify(test_features)

print("Test Sentence:")
print(test_sentence)

print("\nPredicted Sense:")
print(prediction)

# --------------------------------
# Most Informative Features
# --------------------------------

print("\nMost Informative Features:\n")

classifier.show_most_informative_features(5)
```
